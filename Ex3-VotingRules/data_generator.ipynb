```python
from faker import Faker
import csv
from random import randint
import pandas as pd
```

# Big Voting Table

```python
def rand_n(n, min_val, max_val):
    """Génère une liste de n entiers aléatoires entre min_val et max_val."""
    return [randint(min_val, max_val) for _ in range(n)]
```

```python
# Générateur de données en français
fake = Faker('fr_FR')

# Génération de 220 noms
noms = [fake.name() for _ in range(220)]
df = pd.DataFrame({'Nom': noms})
print(df.shape)
df.head() 
```

```python
nb = 20
for i in range(nb):
    name = df.loc[i, 'Nom']
    name = "Cd. " + name
    df[name] = 0

df = df.drop(index=range(0, nb))

print(df.shape)

df.head()
```

```python
v = list(range(1, nb+1))

def permute(lst):
    """Retourne une permutation aléatoire de la liste donnée."""
    lst_copy = lst[:]
    random_permutation = []
    while lst_copy:
        index = randint(0, len(lst_copy) - 1)
        random_permutation.append(lst_copy.pop(index))
    return random_permutation

v = permute(v)

print(v)
```

```python
for i in range(df.shape[0]):
    v = permute(v)
    df.iloc[i, 1:] = v

df.head()
```

```python
df.to_csv('preference.csv', index=False, encoding='utf-8')
print("DataFrame créé et exporté en 'preference.csv'.")
```
